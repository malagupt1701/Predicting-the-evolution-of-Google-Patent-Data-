```python
df_lengths = [1,150456,307779,469536,628516,751018]
outer_list_year = outer_list_pkl[df_lengths[4]-1:df_lengths[5]]
tfidf = models.TfidfModel(outer_list_year)
corpus_tfidf = tfidf[outer_list_year]
```

```python
import pickle
model = pickle.load(open('/content/drive/MyDrive/Capstone/Capstone Project/Capstone Data with claims/lsa_model_2022_new_10_topics.pkl', 'rb'))
corpus_lsi = model[corpus_tfidf]
```

```python
score1 = []
score2 = []
score3 = []
score4 = []
score5 = []
score6 = []
score7 = []
score8 = []
score9 = []
score10 = []
for doc in corpus_lsi:

  try:
    score1.append(round(doc[0][1],4))
    score2.append(round(doc[1][1],4))
    score3.append(round(doc[2][1],4))
    score4.append(round(doc[3][1],4))
    score5.append(round(doc[4][1],4))
    score6.append(round(doc[5][1],4))
    score7.append(round(doc[6][1],4))
    score8.append(round(doc[7][1],4))
    score9.append(round(doc[8][1],4))
    score10.append(round(doc[9][1],4))
  except:
    continue
```

```python
df_topic = pd.DataFrame()
index = [str('topic_' + str(num)) for num in range(10)]
df_topic['topic_0'] = score1
df_topic['topic_1'] = score2
df_topic['topic_2'] = score3
df_topic['topic_3'] = score4
df_topic['topic_4'] = score5
df_topic['topic_5'] = score6
df_topic['topic_6'] = score7
df_topic['topic_7'] = score8
df_topic['topic_8'] = score9
df_topic['topic_9'] = score10

df_topic['Topic']= df_topic.apply(lambda x: x.argmax(), axis=1)
df_topic.head(1)
```

```python
df_topic.to_csv('tsne_2022_lsa.csv',index=False)
```

```python
from sklearn.manifold import TSNE
from bokeh.plotting import figure, output_file, show
from bokeh.models import Label
from bokeh.io import output_notebook

index = [str('topic_' + str(num)) for num in range(10)]
arr=df_topic[index]

# Keep the well separated points (optional)
#arr = arr[np.amax(arr, axis=1) > 0.03]

# Dominant topic number in each doc
topic_num = df_topic['Topic']

# tSNE Dimension Reduction
tsne_model = TSNE(n_components=2, verbose=1, random_state=42, init='random')
tsne_lda2 = tsne_model.fit_transform(arr)
```

```python
import matplotlib.colors as mcolors

# Plot the Topic Clusters using Bokeh
output_notebook()
n_topics = 10
mycolors = np.array([color for name, color in mcolors.TABLEAU_COLORS.items()])
plot = figure(title="t-SNE Clustering of {} LSA Topics".format(n_topics), 
              plot_width=900, plot_height=700)
plot.scatter(x=tsne_lda2[:,0], y=tsne_lda2[:,1], color=mycolors[topic_num])
show(plot)
```
